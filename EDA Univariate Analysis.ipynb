# 1. Categorical Data

## a. CountPLot

```python
# Survived columns is a categorical data because we can categorize it into two.. Either people suvived or not
# Survived = 1,  Not survived = 0
```

```python
sns.countplot(x = 'Survived', data=df, hue='Survived')
# This tells Seaborn:
# x='Survived' → use this column
# data=df → from this DataFrame
# hue='Survived'   this automatically assigns different colours
```

```python
sns.countplot(x='Pclass', data=df, hue = 'Pclass')
```

```python
# TO see data in percentage we can use piechart
df['Pclass'].value_counts()
```

## b. PieChart

```python
"""
PIE CHART (plt.pie) — COMPLETE GUIDE

A pie chart is used to show how a whole (100%) is divided into different
categories. Each slice represents the proportion or percentage of a category.

Best used when:
- Data is categorical
- Number of categories is small (2–5)
- Goal is to show proportion / percentage

Avoid when:
- Too many categories
- Exact comparison is required (use bar chart instead)
"""
```

```python
Pclass_counts = df['Pclass'].value_counts()

# Creates a new figure with width = 6 inches and height = 6 inches
# This controls the overall size of the pie chart
plt.figure(figsize=(6,6))


plt.pie(
    # Data values for the pie chart
    # Each value represents the count of a category
    # Matplotlib converts these values into percentages automatically
    Pclass_counts,

    # Labels for each slice of the pie chart
    # These labels describe what each slice represents
    labels=['3st Class', '1nd Class', '2rd class'],

    # Displays percentage value on each slice
    # %1.1f  -> one digit after decimal
    # %%      -> shows the percent symbol
    autopct='%1.1f%%',

    # Sets custom colors for each slice
    colors=['#ff7675', '#55efc4', 'Green'],

    # Rotates the pie chart by 90 degrees
    # Makes the chart start from the top for better visual balance
    startangle=90,

    # Adds a black border around each slice
    # Helps clearly separate slices for better readability
    wedgeprops={'edgecolor': 'black'}
)

# Adds a title at the top of the chart
plt.title('Titanic Pclass Distribution')

# Displays the pie chart on the screen
plt.show()
```
