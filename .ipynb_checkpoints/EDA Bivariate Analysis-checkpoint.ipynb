```python
# Bivariate Data Analysis (in EDA):
# --------------------------------
# Bivariate analysis means analyzing TWO variables at the same time
# to understand their relationship, pattern, or dependency.
#
# Goal:
# - Check how one variable changes with respect to another
# - Identify correlation, trends, or associations
#
# Common cases:
# 1) Numerical vs Numerical  -> scatter plot, correlation (e.g., height vs weight)
# 2) Categorical vs Numerical -> boxplot, bar plot (e.g., gender vs salary)
# 3) Categorical vs Categorical -> countplot, crosstab (e.g., gender vs survival)
#
# Why it's important:
# - Helps find relationships useful for feature selection
# - Detects patterns, outliers, and trends before modeling
# - Guides better data-driven decisions
#
# Example:
# - Checking how 'Fare' affects 'Survived' in the Titanic dataset
```

# 1. ScatterPlot(Numerical-Numerical)

```python
# Scatter Plot:
# -------------
# A scatter plot is used to visualize the relationship between TWO numerical variables.
# Each dot represents one data point (x-value, y-value).
#
# What it helps to understand:
# - Correlation (positive, negative, or no relationship)
# - Trends or patterns in data
# - Clusters (grouping of points)
# - Outliers (points far from the rest)
#
# When to use:
# - Numerical vs Numerical data
#   e.g., height vs weight, age vs income
#
# How to interpret:
# - Upward trend  -> Positive correlation
# - Downward trend -> Negative correlation
# - Random scatter -> No clear relationship
#
# Commonly used in EDA to:
# - Explore feature relationships
# - Decide useful features for ML models
#
# Example:
# sns.scatterplot(x='Age', y='Fare', data=df)
```

```python
# using tips dataset
sns.scatterplot(x='total_bill', y='tip', data=tips)
```

```python
# Now here lets do it like multivariate analysis side by side
# here we are trying to visualize and understand the relation about 4 columns 
# they are --- total_bills, tip, sex, smoker

# What we can observe:
# - Whether higher bills generally lead to higher tips
# - Differences in tipping behavior by sex
# - Differences between smokers and non-smokers
# - Any visible clusters or outliers
#
# This is a bivariate analysis (total_bill vs tip)
# with categorical grouping for deeper insight.

   
sns.scatterplot(x='total_bill', y='tip', data=tips, hue = "sex", style = 'smoker')
```

# 2. Bar Plot(Numerical_Categorical)

```python
titanic.head(3)
```

```python
# In titanic dataset Pclass is a categorical dataset (1stclass, 2nd class, 3rd class)
# Now I wanna see the age group(Numerical) in each of the class

sns.barplot(x='Pclass', y='Age', data=titanic, hue='Pclass')
```

```python
# Seeing whats the price/fare of each class 
sns.barplot(x='Pclass', y='Fare', data=titanic, hue='Pclass')
```

```python
# It shows that, the price/fare of 1st class is very high compared to others two
```

```python
# Seeing whats the price/fare of each class according to Gender/sex
sns.barplot(x='Pclass', y='Fare', data=titanic, hue='Sex')
```

```python
# Somehow females have to pay higher prices/fares than males
```
