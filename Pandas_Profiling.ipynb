```python
# Pandas Profiling (ydata-profiling) automatically generates a detailed
# Exploratory Data Analysis (EDA) report for a pandas DataFrame.
# It shows dataset overview, column-wise statistics, missing values,
# distributions, correlations, and data quality warnings.
# This helps quickly understand and clean data before ML modeling.
```

```python
import pandas as pd
```

```python
df = pd.read_csv('Titanic-Dataset.csv')
```

```python
df.head()
```

```python
# Import ProfileReport class from ydata_profiling for automatic EDA
from ydata_profiling import ProfileReport

# Create a profiling report for the DataFrame `df`
# title="EDA Report" sets the report title
# explorative=True enables advanced analysis and embeds all JS/CSS
# so that the saved HTML remains fully interactive (tabs, alerts, plots)
profile = ProfileReport(
    df,
    title="EDA Report",
    explorative=True
)

# Save the profiling report as a self-contained interactive HTML file
# which can be opened in any web browser
profile.to_file("eda_report.html")
```
